```python
import matplotlib.pyplot as plt
import math
from project import compute_ap
from tqdm import tqdm
```

```python
# 2. Sieve of Eratosthenes up to M
M = 60000
sieve = [True] * (M + 1)
sieve[0] = sieve[1] = False

for p in range(2, int(math.isqrt(M)) + 1):
    if sieve[p]:
        for multiple in range(p*p, M+1, p):
            sieve[multiple] = False

# 3. Collect primes
primes = [i for i, is_prime in enumerate(sieve) if is_prime]
primes = primes[:5000]
```

```python
print(len(primes))
print(primes[-1])
```

```python
params = [(1,1),(0,1),(2,1),(-3,2)]
for a, b in params:
    tp = []
    for i in tqdm(range(len(primes))):
        p = primes[i]
        ap = compute_ap(p, a, b)
        tp.append(ap /(2 * math.sqrt(p)))
    plt.figure()
    plt.hist(tp, bins=70)
    plt.title(f"Histogram for (a,b) = ({a}, {b})")
    plt.ylabel("frequency")
    plt.xlabel("t_p")
    plt.show()
```

```python
# p equiv 1, -1 mod 3
prime_one = []
prime_negative_one = []

for p in primes:
    if p % 3 == 1:
        prime_one.append(p)
    #
    elif p % 3 == 2:
        prime_negative_one.append(p)
```

```python
def get_primes():
    M = 60000
    sieve = [True] * (M + 1)
    sieve[0] = sieve[1] = False

    # isqrt(M) returns the biggest integer n such that n^2 <= M. We only need to check up to the sqrt  
    for p in range(2, int(math.isqrt(M)) + 1):
        if sieve[p]:
            
            # cross out all multiples of the current number
            for multiple in range(p*p, M+1, p):
                sieve[multiple] = False

    # 3. Collect primes
    primes = [i for i, is_prime in enumerate(sieve) if is_prime]
    return primes[:5000]
```

```python
def problem_7():
    # p equiv 1, -1 mod 3
    primes = get_primes()
    prime_one = []
    prime_negative_one = []

    for p in primes:
        if p % 3 == 1:
            prime_one.append(p)
        #
        elif p % 3 == 2:
            prime_negative_one.append(p)


    tp = []
    for i in range(len(prime_one)):
        p = prime_one[i]
        ap = compute_ap(p, 0, 1)
        tp.append(ap /(2 * math.sqrt(p)))
    plt.figure()
    plt.hist(tp, bins=70)
    plt.title(f"Histogram for (a,b) = ({0}, {1}), p = 1 (mod 3)")
    plt.ylabel("frequency")
    plt.xlabel("t_p")
    plt.show()

    tp = []
    for i in range(len(prime_negative_one)):
        p = prime_negative_one[i]
        ap = compute_ap(p, 0, 1)
        tp.append(ap /(2 * math.sqrt(p)))
    plt.figure()
    plt.hist(tp, bins=70)
    plt.title(f"Histogram for (a,b) = ({0}, {1}), p = -1 (mod 3)")
    plt.ylabel("frequency")
    plt.xlabel("t_p")
    plt.show()
```
